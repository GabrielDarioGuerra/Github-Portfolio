```python
import seaborn as sns
import numpy as np

# Removing 'id' column
numerical_columns = df_average.select_dtypes(include=['int64', 'float64']).columns.drop('id')

# Calculating correlation matrix
correlation_matrix = df_average[numerical_columns].corr()

# Creating a mask to hide the diagonal values
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Plotting heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", mask=mask)
plt.title('Correlation Heatmap of Numerical Values (excluding "id" and diagonal)')
plt.show()
```

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Removing 'id' column
numerical_columns = df_average.select_dtypes(include=['int64', 'float64']).columns.drop('id')

# Calculating correlation matrix
correlation_matrix = df_average[numerical_columns].corr()

# Creating a mask to hide the diagonal values
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# Plotting heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", mask=mask)

# Adjusting tick parameters to hide empty fields
plt.xticks(rotation=45)
plt.yticks(rotation=0)

plt.title('Correlation Heatmap of Numerical Values (excluding "id" and diagonal)')
plt.show()
```
